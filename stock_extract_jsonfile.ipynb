```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
import json
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
import time
import re
```

```python
with open("/Users/elaine/Desktop/side_project/stock_price/3711.json","r") as f:  #change file
    
    json_data = json.loads(f.read())
    json_str = '[' + ','.join(json_data) + ']'
df = pd.read_json(json_str)
df
```

```python
df = df.explode('data')
df = pd.concat([df.drop(['data'], axis=1), df['data'].apply(pd.Series)], axis=1) #axis = 0(直向合併) axis = 1(左右合併)
df.drop(df.columns[0:3],axis=1,inplace = True)

df
```

```python
df.drop(df.columns[1:3],axis=1,inplace = True)
df
```

```python
df.columns = ["Stock Code","Date","Trade Volume","Trade Value","Opening Price","Highest Price","Lowest Price","Closing Price","Change","Transaction"]
df
```

```python
df["Date"]
```

```python
Company_list=[]
for stock_num in df["Stock Code"]:
    if stock_num == None:
        continue
    if isinstance(stock_num,str):
        code = stock_num[7:12]
    Company_list.append(code)
#print(Company_list)       

df["Stock Code"] = Company_list
df
```
